```python
# https://scikit-learn.org/stable/auto_examples/svm/plot_svm_kernels.html
def plotSVM(degree,gamma):
    clf = svm.SVC(kernel='poly', degree=degree, gamma=gamma)
    clf.fit(X, Y)

    # plot the line, the points, and the nearest vectors to the plane
    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80, facecolors='none', zorder=10, edgecolors='k')
    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired, edgecolors='k')
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(XX.shape)
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],levels=[-.5, 0, .5])
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.title(f"Degree={degree}, Gamma={gamma}")
    plt.show()
plotSVM(2,1)
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor

# Load the Iris dataset
iris = datasets.load_iris(as_frame=True)

# print part of the dataset
iris_X, iris_y = iris.data, iris.target

# Plot also the training points
p1 = plt.scatter(iris_X.iloc[:, 0], iris_X.iloc[:, 1], c=iris_y, s=50, cmap=plt.cm.Paired)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(handles=p1.legend_elements()[0], labels=['Setosa', 'Versicolour', 'Virginica', 'Query'], loc='lower right')
```

```python
from sklearn.tree import DecisionTreeClassifier
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

def make_grid(X):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X.iloc[:, 0].min() - 0.1, X.iloc[:, 0].max() + 0.1
    y_min, y_max = X.iloc[:, 1].min() - 0.1, X.iloc[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    return xx, yy, x_min, x_max, y_min, y_max

def make_2d_preds(clf, X):
    xx, yy, x_min, x_max, y_min, y_max = make_grid(X)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    return Z

def make_2d_plot(ax, Z, X, y,depth):
    # Create color maps
    cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
    cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

    xx, yy, x_min, x_max, y_min, y_max = make_grid(X)

    # Put the result into a color plot
    ax.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

    # Plot also the training points
    ax.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap=plt.cm.Paired, edgecolor='k', s=50)
    ax.set_xlabel('Sepal Length')
    ax.set_ylabel('Sepal Width')
    ax.set_title(f'Max_Depth={depth}')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())   
    plt.show()
```

```python
# 2.1 Homogeneous Decision Tree, followed by 3 depths, 2, 4, 8

X = iris_X.iloc[:,:2]
max_depths=[None,2,4,8]
for i in max_depths:
    ax = plt.gca()
    clf = DecisionTreeClassifier(max_depth=i)
    clf.fit(X, iris_y)
    Z = make_2d_preds(clf, X)
    make_2d_plot(ax, Z, X, iris_y,i)
```

```python
#Q2.3, plot 1. Depth=None, Trees= 10,Num samples=20
def RandomForestClassification(num_trees,num_samples,max_depth):
    np.random.seed(1000)
    n_models, n_flowers, n_subset = num_trees, iris_X.shape[0], num_samples
    random_forest = []
    for i in range(n_models):
        # sample the data with replacement
        random_idx = np.random.randint(0, n_flowers, size=(n_subset,))
        X_random, y_random = iris_X.iloc[random_idx, :2], iris_y[random_idx]

        # train a decision tree model
        clf = DecisionTreeClassifier(max_depth=max_depth)
        clf.fit(X_random, y_random)

        # append it to our ensemble
        random_forest += [clf]

    # Visualize predictions from a random forest
    ax = plt.gca()

    # compute average predictions from all the models in the ensemble
    X_all, y_all = iris_X.iloc[:,:2], iris_y
    Z_list = []
    for clf in random_forest:
        Z_clf = make_2d_preds(clf, X_all)
        Z_list += [Z_clf]
    Z_avg = np.stack(Z_list, axis=2).mean(axis=2)

    # visualize predictions
    make_2d_plot(ax, np.rint(Z_avg), X_all, y_all,max_depth)
```

## Question 3: (40) (Lecture 12: Decision tree for linear regression)

In the lecture note, we have performed decision tree for classification on Iris data set. We have learned that if we use a large depth the decision tree will overfit the data, leading to predictions with high variance. Bagging is a technique to reduce high variance. 

Question: 
1. Similarly to the logistic regression problem in the Lecture note 12, you can use the given code below, and then perform bagging method (with max_depth = 5) to reduce the overfitting. You should test a few cases: 
    a. different number of random trees, 
    b. different sizes for random subsets of the data. 

2. Discuss the result.

```python
# Import the necessary modules and libraries
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))

# Fit regression model
regr = DecisionTreeRegressor(max_depth=5)
regr.fit(X,y)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_test = regr.predict(X_test)


# Plot the resultsPP
plt.figure(figsize=(10,8))
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X_test, y_test, color="yellowgreen", label="Overfitted Decision Tree", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Overfitted Decision Tree")
plt.legend()
plt.show()
```

```python
#Defining random forest function with a sample plot:
plt.figure(figsize=(10,8))
def RandomForestRegression(sample_sizes,num_trees):
    y_test=[]
    for i in range(num_trees):
        random_idx = np.random.randint(0, 80, size=(sample_sizes,))
        X_random, y_random = X[random_idx], y[random_idx]
        regr.fit(X_random,y_random)
        y_test.append(regr.predict(X_test))

    y_test_random=np.mean(y_test,axis=0)
    plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
    plt.plot(X_test, y_test_random, color="yellowgreen", label="Random Forest", linewidth=2)
    plt.xlabel("data")
    plt.ylabel("target")
    plt.title(f"Samples={sample_sizes}, Number of trees={num_trees}")
    plt.legend()
    plt.show()
RandomForestRegression(10,300)
```

```python
sample_sizes=np.array([5,10,20,30,50,80])
num_trees=np.array([10,20,50,100,200])
for i in sample_sizes:
    for j in num_trees:
        RandomForestRegression(i,j)
```

It can be observed that the Random Forest (Bagging Method) works well in reducing overfitting of data. In particular, smaller sample sizes (very small like 5 leads to underfitting) work well in reducing overfitting as compared to larger sample sizes, and using more number of trees efficiently reduces overfitting (Till a certain point, and then it saturates). A good strategy found for this problem was to use a relatively smaller sample size (10-20) with higher number of trees (100-200).
